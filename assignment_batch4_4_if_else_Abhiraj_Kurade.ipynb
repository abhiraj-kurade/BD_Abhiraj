### 36. Determine the type of triangle using sides: acute, right, obtuse.

```python
a1 = 90
a2 = 60
a3 = 30

if a1 + a2 + a3 == 180 and a1>0 and a2>0 and a3>0:
    if a1 == a2 == a3:
        print("Equilateral triangle")
    elif a1 == 90 or a2 == 90 or a3 == 90:
        print("Right angled triangle")
    elif a1 < 90 and a2 < 90 and a3 < 90:
        print("Acute angled triangle")
    elif a1 > 90 or a2 > 90 or a3 > 90:
        print("Obtuse angled triangle")
    else:
        print("Scalene triangle")
else:
    print("Not a valid triangle")
```

### 48. Check if a password contains both letters and digits.

```python
password = input("Enter password: ") 

digit = False
alpha = False

for ch in password: 
    if ch.isdigit():
        digit = True 
    if ch.isalpha():
        alpha = True 

if digit and alpha: 
    print("Password contain both alphabets and digits")
else:
    print("Password does not contain both alphabets and digits")
```
