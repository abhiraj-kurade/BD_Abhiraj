```python
# 1. LIST OPERATIONS (Mutable)

fruits = ["apple", "banana", "cherry"]
```

```python
#Access Elements

print(fruits[0])      # apple
print(fruits[-1])     # cherry
```

```python
#Add Elements

fruits.append("orange")        # Add at end
fruits.insert(3, "mango")      # Insert at index
print(fruits)
```

```python
#Remove Elements

fruits.remove("banana")   # remove by value
fruits.pop(1)             # remove by index
fruits.pop()              # remove last element
del fruits[0]             # delete element
```

```python
#Update Elements

fruits[0] = "kiwi"      # update index 0     
print(fruits)
```

```python
#List Slicing

print(fruits[1:3])
```

```python
#List Methods

numbers = [3, 1, 4, 2]
numbers.sort()            # [1, 2, 3, 4]
numbers.reverse()         # reverse list
print(len(numbers))       # length
```

```python
# TUPLE OPERATIONS (Immutable)

t = (10, 20, 30)
```

```python
#Access Elements

print(t[0])     # 10
```

```python
#Count & Index Methods


t = (1, 2, 2, 3)
print(t.count(2))   # 2
print(t.index(3))   # 3rd element index


```

```python
#SET OPERATIONS (Unordered, no duplicates)


s = {1, 2, 3, 3}
print(s)   # {1, 2, 3}
```

```python
#Add / Remove

s.add(4)
s.remove(2)
s.discard(5)    # No error if not found
```

```python
#Set Operations


a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))          # {1,2,3,4,5}
print(a.intersection(b))   # {3}
print(a.difference(b))     # {1,2}
print(a.symmetric_difference(b)) # {1,2,4,5}
```

```python
#Check Membership

print(5 in a)

print(1 in a)
```

```python
#DICTIONARY OPERATIONS (key-value pairs)

student = {
    "name": "Abhiraj",
    "age": 24,
    "course": "BigData"      #creating a dictionary
}
```

```python
#Access Values

print(student["name"])
print(student.get("course"))
print(student.get("age"))

```

```python
#Add / Update
    
student["age"] = 25
student["city"] = "Mumbai"
```

```python
#Delete

student.pop("city")
del student["course"]
```

```python
#Dictionary Methods


print(student.keys())
print(student.values())
print(student.items())
```
