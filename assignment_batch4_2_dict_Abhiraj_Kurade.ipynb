# Basic Dictionay Operations 

1. Create a dictionary with 3 key-value pairs (name, age, city).


```python
person = {"name": "Abhiraj", "age": 25, "city": "Navi Mumbai"} 
print(person)
```

2. Create a dictionary and print the value of a specific key.


```python
person1 = {"name": "Harish", "age": 26, "city": "Pune"} 
print(person1["age"]) 
```

3. Make a dictionary and add a new key-value pair to it.


```python
person3 = {"name": "Jethalal", "age": 40} 
person3["city"] = "Mumbai" 
print(person3)
```

4. Create a dictionary and change the value of an existing key.


```python
person4 = {"name": "Jethalal", "age": 40}  
person4["age"] = "41"  
print(person4)
```

5. Create a dictionary and remove one key using pop().


```python
person5 = {"name": "Abhiraj Kurade", "age": 25, "city": "Navi Mumbai"} 
person5.pop("age") 
print(person5)
```

6. Create a dictionary and remove the last inserted item using popitem().


```python
person6 = {"name": "Ashu", "age": 24, "city": "Kokhapur"} 
person6.popitem() 
print(person6)
```

7. Create a dictionary and print all the keys.


```python
person7 = {"name": "Bobby", "age": 22, "city": "Kolhapur"} 
print(person7.keys())
```

8. Create a dictionary and print all the values.


```python
person8 = {"name": "Bobby", "age": 22, "city": "Kolhapur"} 
print(person8.values())
```

9. Create a dictionary and print all the key-value items.


```python
person9 = {"name": "Omkar", "age": 28, "city": "Pune"} 
print(person9.items()) 
```

10. Make a dictionary and check the number of items using len().


```python
person10 = {"name": "Jackie", "age": 7, "city": "Kolhapur"} 
print(len(person10))
```

11. Create a dictionary and check if a certain key exists (using "key" in dict).


```python
person11 = {"name": "Raj", "age": 33, "city": "Delhi"} 

print("country" in person11)  # False 

print("age" in person11)  # True 
```

13. Create a dictionary and update one keyâ€™s value using update().


```python
person13 = {"name": "Kshitij", "age": 22} 
person13.update({"age": 24}) 
print(person13)
```

14. Create two dictionaries and combine them using update().


```python
dict1 = {"name": "Rohit45", "age": 38} 
dict2 = {"city": "Mumbai", "country": "India"} 
dict1.update(dict2)
print(dict1)
```

15. Make a dictionary and get the value of a key using get().


```python
person15 = {"name": "Abhiiraaj", "age": 26, "city": "Pune"} 
print(person15.get("city")) 
print(person15.get("country", "Not Found"))  # it will return 'Not Found' as 'country' key is not present in the dictionary
```
