## **3. Count how many even numbers are present in a nested list.**

````python
# Write your answer here
lst2 = [[10,20], [30], [5, 7, 8]]
count = 0

for i in lst2:
    for j in i:
        if j %2 == 0:
            count +=1
print(count)
````

## **4. Convert all strings to uppercase inside a nested list.**
Example: `[['hello', 'world'], ['python']]`

````python
# Write your answer here
lst4 = [['hello', 'world'], ['python']]
for i in range(len(lst4)):
    for j in range(len(lst4[i])):
        lst4[i][j] = lst4[i][j].upper()
print(lst4)
````

## **6. Create a new list containing numbers greater than 50 from a nested list.**

````python
# Write your answer here
lst = [[10, 60, 55], [20, 90], [51, 30]]

result = []

for i in lst:
    for j in i:
        if j > 50:
            result.append(j)

print(result)
````

## **7. Find the largest number inside a nested list.**

````python
# Write your answer here
lst7= [[10,20], [50, 5], [7, 99]]

max = 0

for i in lst7:
    for j in i:
        if j > max:
            max = j

print(max)
````

## **8. Flatten a nested list into a single list.**
`[[1,2], [3,4], [5]] â†’ [1,2,3,4,5]`

````python
# Write your answer here
lst = [[1,2], [3,4], [5]]

flat = []

for sub in lst:
    for num in sub:
        flat.append(num)
print(flat)
````
