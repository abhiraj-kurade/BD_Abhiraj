## **4. Convert all strings to uppercase inside a nested list.**
Example: `[['hello', 'world'], ['python']]`

````python
# Write your answer here
lst4 = [['hello', 'world'], ['python']]
for i in range(len(lst4)):
    for j in range(len(lst4[i])):
        lst4[i][j] = lst4[i][j].upper()
print(lst4)
````

## **8. Flatten a nested list into a single list.**
`[[1,2], [3,4], [5]] â†’ [1,2,3,4,5]`

````python
# Write your answer here
lst = [[1,2], [3,4], [5]]

flat = []

for sub in lst:
    for num in sub:
        flat.append(num)
print(flat)
````

## **9. Replace all negative numbers with 0 inside a nested list.**

````python
# Write your answer here
lst9=[[80, -62, 85], [-80, 70, 35], [49, 92, -65]]
for i in range (len(lst9)): #outer list 
    for j in range (len(lst9[i])):  #inner list value
        if(lst9[i][j]<0):    #outer inner  < zero condition
            lst9[i][j]=0    # less than zero then replace with zero 
print(lst9)
````

## **20. Convert all inner dictionary keys to uppercase.**

````python
# Write your answer here
data = {
    'A': {'age': 15, 'grade': 10},
    'B': {'age': 16, 'grade': 11}
}
new_data = {}
for i, j in data.items():
    new_data[i] = {}
    for key, value in j.items():
        new_data[i][key.upper()] = value

print(new_data)
````
